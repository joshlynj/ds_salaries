```python
# highest_pay = ds_df.groupby("job_title")["salary_in_usd"].max().sort_values(ascending=False)
# highest_pay_top = highest_pay.head(10).sort_values(ascending=False).to_frame()
# highest_pay_top


highest_pay = ds_df.groupby("job_title")["salary_in_usd", "employment_type", "company_location"].min().sort_values(by = 'salary_in_usd', ascending=False)
#lowest_pay_top = lowest_pay.head(10).sort_values(ascending=False).to_frame()
mask = highest_pay['company_location'].isin(["US"]) & highest_pay['employment_type'].isin(["FT"])
highest_pay_top = highest_pay[mask].head(10)
highest_pay_top
```

```python
fig, ax1 = plt.subplots(figsize=(16,9))
sns.barplot(y = highest_pay_top.index, x = highest_pay_top.salary_in_usd, orient="h", palette="crest")
# ax.set_xlim(100000, 460000)
plt.xlabel('Salary (in USD)')
plt.ylabel('Job Title')
plt.title('The 10 Highest Paid Data Science Jobs In The U.S. (2022-2023)')
```

```python
ds_df
ft_only_df = ds_df.drop(ds_df[ds_df['employment_type'] != "FT"].index)
us_ft_df = ft_only_df.drop(ft_only_df[ft_only_df['company_location'] !="US"].index)
us_ft_df
```

```python
us_ft_df['job_count'] = us_ft_df.groupby('job_title')['job_title'].transform('count')
us_ft_df
```

```python
n = 10
most_common = us_ft_df['job_title'].value_counts().head(10).to_frame()
# most_common = ds_df["job_title"].mode()
display(most_common)

```

```python
most_common_df = us_ft_df.groupby("job_count", as_index=False)[["salary_in_usd", "job_title"]].mean(numeric_only=True).sort_values(by = 'job_count', ascending=False).head(10)


index_labels = ["Data Engineer", 
         "Data Scientist", 
         "Data Analyst", 
         "Machine Learning Engineer", 
         "Data Architect",
         "Analytics Engineer", 
         "Applied Scientist", 
         "Research Scientist", 
         "Data Science Manager", 
         "Research Engineer"]

most_common_df.index = index_labels
display(most_common_df)
```

```python
fig, ax1 = plt.subplots(figsize=(16,9))
sns.barplot(x = most_common_df.index, y = most_common_df.salary_in_usd, palette="crest")
ax2=ax1.twinx()
sns.lineplot(data = most_common_df['job_count'], marker='o', sort = False, ax=ax2, color = '#aa42f5')
ax1.set_ylabel('Salary (in USD)')
ax2.set_ylabel('Job Count', color = '#aa42f5')
ax1.set_xlabel('Job Title')
ax1.tick_params(axis='x', rotation=45)
plt.title('The 10 Most Common Jobs and Their Average Pay (2022-2023)')
```

```python
remote_df = us_ft_df[['remote_ratio',"company_size"]].sort_values(by='company_size', ascending=False)
#type(remote_df)
remote_df
```

```python
grouped = remote_df.groupby(['company_size', 'remote_ratio']).size().unstack()

grouped.plot(kind='bar', stacked=True)
plt.xlabel('Company Size')
plt.ylabel('Count')
plt.title('Remote Ratios by Company Size')
plt.legend(title='Remote Ratio', loc='upper right')
plt.show()
```

```python
s_df = remote_df[remote_df['company_size'] == 'S']
m_df = remote_df[remote_df['company_size'] == 'M']
l_df = remote_df[remote_df['company_size'] == 'L']

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16, 6))

sns.countplot(x='remote_ratio', data=s_df, ax=axes[0], palette="crest")
axes[0].set_title('Company Size S')
axes[0].set_xlabel('Remote Ratio')
axes[0].set_ylabel('Count')

sns.countplot(x='remote_ratio', data=m_df, ax=axes[1], palette="crest")
axes[1].set_title('Company Size M')
axes[1].set_xlabel('Remote Ratio')
axes[1].set_ylabel('Count')

sns.countplot(x='remote_ratio', data=l_df, ax=axes[2], palette="crest")
axes[2].set_title('Company Size L')
axes[2].set_xlabel('Remote Ratio')
axes[2].set_ylabel('Count')

plt.suptitle('Remote Ratios by Company Size', fontsize=16)

plt.tight_layout()
```
